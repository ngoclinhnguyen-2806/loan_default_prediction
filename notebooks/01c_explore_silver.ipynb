## Test gold feature builder

```python
import sys
from pyspark.sql import SparkSession, functions as F
from pathlib import Path

sys.path.append("/app/utils")
from gold_features_processing_optimized import _load_table, make_training_apps, build_features

# ---------------------------------------------------------------
# 1. Spark Setup
# ---------------------------------------------------------------
spark = (
    SparkSession.builder
    .appName("test_gold_builder")
    .master("local[*]")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("ERROR")

silver_dir = "/app/datamart/silver"

print("\n=== üîç Checking Silver Layer Tables ===")
for tbl in ["lms_loan_daily", "features_financials", "features_attributes", "feature_clickstream"]:
    df = _load_table(spark, silver_dir, tbl)
    if df:
        print(f"‚úÖ {tbl}: {df.count():,} rows, {len(df.columns)} cols")
        df.show(3, truncate=False)
    else:
        print(f"‚ö†Ô∏è  {tbl}: Not found or empty")

# ---------------------------------------------------------------
# 2. Prepare mock training applications
# ---------------------------------------------------------------
lms_df = _load_table(spark, silver_dir, "lms_loan_daily")
asof_date = "2024-12-31"

if lms_df:
    apps_train = make_training_apps(lms_df, asof_date)
    print(f"\n=== üß© Training applications prepared ===")
    apps_train.show(5, truncate=False)

    # ---------------------------------------------------------------
    # 3. Build features from Silver layer
    # ---------------------------------------------------------------
    gold_df = build_features(apps_train, silver_dir, asof_date, spark)

    print("\n=== ‚ú® Gold Feature Summary ===")
    print(f"Total rows: {gold_df.count():,}")
    print(f"Total columns: {len(gold_df.columns)}")

    # Show sample columns and stats
    gold_df.select(
        "Customer_ID", "application_date", "Annual_Income",
        "Outstanding_Debt", "DTI", "estimated_EMI", "EMI_to_income"
    ).show(5, truncate=False)
else:
    print("‚ö†Ô∏è  Skipping feature build ‚Äî missing lms_loan_daily")

# ---------------------------------------------------------------
# 4. Cleanup
# ---------------------------------------------------------------
spark.stop()
```
